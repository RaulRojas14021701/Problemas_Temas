Problema de la venta de periodicos

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def periodicos():
    vende=0
    x=random.random()
    if(x<=0.20):
          vende=30
    if(x<=0.45 and x>0.20):
	        vende=40
    if(x<=0.70 and x>0.45):
	        vende=50
    if(x<=0.90 and x>0.70):
	        vende=60
    if(x<=1.0 and x>0.90):
	        vende=70
    return vende
```

```python
def ganancia():
	gana=0
	vendio=periodicos()
	if vendio==50:
		gana=0.2*50
	if vendio<50:
		gana=(0.2*(vendio))-(0.1*(50-vendio))
	if vendio>50:
		gana=(0.2*(50))-(0.2*(vendio-50))
	return gana
```

```python
def promedia(n):
    prom=[]
    tiros=0
    for i in range(n):
        tiros+=ganancia()
        prom.append(tiros/(i+1))
    return prom
```

```python
X=promedia(400)
```
